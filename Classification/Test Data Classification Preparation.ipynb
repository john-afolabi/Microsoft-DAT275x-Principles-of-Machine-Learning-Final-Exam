```python
import pandas as pd
import numpy as np
from datetime import datetime,time,timedelta,date
```

```python
AW_test = pd.read_csv('AW_test.csv')
```

```python
AW_test.head()
```

```python
AW_test.drop(['Title','FirstName','LastName','MiddleName','Suffix','AddressLine1',
           'AddressLine2','StateProvinceName','PhoneNumber','City', 'PostalCode'], axis=1, inplace=True)
```

```python
AW_test.dtypes
```

```python
AW_test['BirthDate'] =  pd.to_datetime(AW_test['BirthDate'])
AW_test['Coll_date']=date(year = 1998, month = 1, day = 1)
AW_test['Coll_date']=pd.to_datetime(AW_test['Coll_date'])
AW_test['Age_days'] = AW_test['Coll_date']- AW_test['BirthDate']
AW_test['Age'] =  AW_test['Age_days'].astype('timedelta64[Y]')
AW_test.drop(['Coll_date','Age_days', 'BirthDate'], axis=1, inplace=True)
AW_test['Age'] = AW_test['Age'].astype(int)
```

```python
AW_test.head()
```

```python
AW_test['AgeRange'] = ['Under 25' if ((x < 25)) 
                   else 'Between 25 and 45' if ((x>=25 and x<=45)) 
                   else 'Between 45 and 55' if ((x>45  and x<=55))
                   else 'Over 55' if ((x>55))
                   else x for x in AW_test['Age']]
```

```python
AW_test['log_YearlyIncome'] = np.log(AW_test['YearlyIncome'])
AW_test.drop(['NumberChildrenAtHome','Age', 'HomeOwnerFlag', 'YearlyIncome'], 
          axis=1, inplace=True)
```

```python
AW_test.to_csv('AW_Test_prepped_classification.csv', index = False, header = True)
```
